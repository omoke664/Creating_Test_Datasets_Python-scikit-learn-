**Create Test DataSets Using Sklearn**

```python
#importing libraries
from sklearn.datasets import make_blobs
#matplotlib for plotting
from matplotlib import pyplot as plt
from matplotlib import style
```

**sklearn.datasets.makeblobs**

```python
#creating test datasets using sklearn.datasets.make_blobs
style.use('fivethirtyeight')
x, y = make_blobs(n_samples = 100, centers = 3,
                  cluster_std = 1, n_features = 2)
plt.scatter(x[:, 0], x[:, 1], s = 40, color = 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
plt.clf()
```

**sklearn.datasets.make_moon**

```python
#creating test datasets using sklearn.datasets.make_moon
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import matplotlib.style as style

#generating the dataset using make_moons
x, y = make_moons(n_samples = 1000, noise = 0.1) #setting the number of data points and the standard deviation.
plt.scatter(x[:,0],x[:,1], s = 40, color = 'g') #selecting columns for coordinates and setting the size and color of plots

#setting the x and y label for the plots
plt.xlabel('x')
plt.ylabel('y')
#display the plot
plt.show()
#generating a clear figure
plt.clf()
```

**sklearn.datasets.make_circle**

```python
#creating test datasets using sklearn.datasets.make_circles
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('fivethirtyeight')
#generating the make circles dataset 
x, y = make_circles(n_samples = 100, noise  = 0.1) #100 datapoints and a variance of 0.1
plt.scatter(x[:,0,],x[:,1], s= 40, color = 'b')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
plt.clf()
```
